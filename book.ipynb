### Шаг 3. Формулировка ML-задачи на основе бизнес-задачи

создадим новые колонки, в которых будут значения цены обслуживания и стоимости номера за ночь с учета сезона

```python
df
```

```python
def months(m):
    if (m in ['January', 'February', 'December']):
        return 1
    if (m in ['March', 'April', 'May']):
        return 1.2
    if (m in ['June', 'July', 'August']):
        return 1.4
    if (m in ['September', 'October', 'November']):
        return 1.2

df['cf_months'] = df.arrival_date_month.apply(months)
```

```python
df.cf_months.unique()
```

```python
df.reserved_room_type = df.reserved_room_type.apply(lambda x: x.split(' ')[0])
print(df.reserved_room_type.unique())
```

```python
df.meal = df.meal.apply(lambda x: x.split()[0])
df.meal.unique()
```

```python
def price(x):
    if (x.is_canceled == 1):
        if (x.reserved_room_type == 'A'):
            return -(1400 * x.cf_months)
        if (x.reserved_room_type == 'B'):
            return -(1150 * x.cf_months)
        if (x.reserved_room_type == 'C'):
            return -(950 * x.cf_months)
        if (x.reserved_room_type == 'D'):
            return -(700 * x.cf_months)
        if (x.reserved_room_type == 'E'):
            return -(650 * x.cf_months)
        if (x.reserved_room_type == 'F'):
            return -(600 * x.cf_months)
        if (x.reserved_room_type == 'G'):
            return -(500 * x.cf_months)
    else:
        if (x.reserved_room_type == 'A'):
            return (x.total_nights * 1000 - (x.total_nights // 2 + 1) * 400) * x.cf_months
        if (x.reserved_room_type == 'B'):
            return (x.total_nights * 800 - (x.total_nights // 2 + 1) * 350) * x.cf_months
        if (x.reserved_room_type == 'C'):
            return (x.total_nights * 600 - (x.total_nights // 2 + 1) * 350) * x.cf_months
        if (x.reserved_room_type == 'D'):
            return (x.total_nights * 550 - (x.total_nights // 2 + 1) * 150) * x.cf_months
        if (x.reserved_room_type == 'E'):
            return (x.total_nights * 500 - (x.total_nights // 2 + 1) * 150) * x.cf_months
        if (x.reserved_room_type == 'F'):
                return (x.total_nights * 450 - (x.total_nights // 2 + 1) * 150) * x.cf_months
        if (x.reserved_room_type == 'G'):
                return (x.total_nights * 350 - (x.total_nights // 2 + 1) * 150) * x.cf_months




df['profit'] = df.apply(lambda x: price(x), axis=1)
df
```

### Шаг 4. Разработка модели ML

```python
def str_to_int_months(x):
    if x == 'January':
        return 1
    if x == 'February':
        return 2
    if x == 'March':
        return 3
    if x == 'April':
        return 4
    if x == 'May':
        return 5
    if x == 'June':
        return 6
    if x == 'July':
        return 7
    if x == 'August':
        return 8
    if x == 'September':
        return 9
    if x == 'October':
        return 10
    if x == 'November':
        return 11
    if x == 'December':
        return 12

df.arrival_date_month = df.arrival_date_month.apply(str_to_int_months)
```

нормируем

```python
df_dummies = pd.get_dummies(df, columns=['meal', 'country', 'distribution_channel', 'reserved_room_type', 'customer_type'], drop_first=True)
df_dummies
```

```python
df_train_dummies = df_dummies.query('is_test == 0')
df_test_dummies = df_dummies.query('is_test == 1')
print(df_test_dummies.shape)
```

```python
features_train = df_train_dummies.drop(['id', 'is_canceled', 'cf_months', 'profit'], axis=1)
target_train = df_train_dummies['is_canceled']

features_test = df_test_dummies.drop(['id', 'is_canceled', 'cf_months', 'profit'], axis=1)
target_test = df_test_dummies['is_canceled']
```

```python
features_train.head()
```

масштабируем для выборок отдельно

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(features_train)
features_train_scaler = pd.DataFrame(data=scaler.transform(features_train), columns=features_train.columns)

scaler.fit(features_test)
features_test_scaler = pd.DataFrame(data=scaler.transform(features_test), columns=features_test.columns)
```

```python
features_test_scaler
```

```python
target_train.hist(bins=100)
```

Перебираем модели

Дерево решений

```python
%matplotlib
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

res = 0
max_i = 0
for i in range(1, 20):
    model = DecisionTreeClassifier(random_state=12345, max_depth=i, class_weight='balanced')
    cvs_model = cross_val_score(model, features_train_scaler, target_train, cv=8, scoring='roc_auc')
    if (cvs_model.mean() > res):
        max_i = i
        res = cvs_model.mean()

print(res, max_i)
```

лучшее значение глубины = 2, roc_auc = 0.7776204125095799

Лес решений

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

parametrs = {
                'n_estimators': [60, 120, 180, 240], 
                'max_depth': [2, 4, 6, 8, 10],
                'max_features': [2, 3, 4, 5]
            }

model = RandomForestClassifier(random_state=12345, class_weight='balanced')
cLf_model = GridSearchCV(model, parametrs, scoring= 'roc_auc', cv=8, n_jobs=-1)
cLf_model.fit(features_train_scaler, target_train)

print(cLf_model.best_score_, cLf_model.best_params_)
```

лучшие значения =  {'max_depth': 2, 'max_features': 4, 'n_estimators': 60}, roc_auc = 0.7686945824203684

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

features_train_test, features_valid_test, target_train_test, target_valid_test = \
            train_test_split(features_train_scaler, target_train, test_size=0.25, random_state=12345, shuffle=True, stratify=target_train)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=12345, class_weight='balanced', solver='lbfgs', max_iter=1000)
model.fit(features_train_test, target_train_test)
predict_proba = model.predict_proba(features_valid_test)

probabilities_one_valid = [i[1] for i in predict_proba] # оставляем только положительную вероятность

f1 = 0
trash = 0

for threshold in np.arange(0, 1, 0.02): # ГРАНИЦЫ ПОРОГА
    predicted_valid = probabilities_one_valid > threshold # булев массив с разными значениями порога
    if f1 < f1_score(target_valid_test, predicted_valid):
        f1 = f1_score(target_valid_test, predicted_valid)
        trash = threshold

print(f1, trash)
```

Порог можно не менять, он точности не добавляет

рассчитаем доход с помощью модели:

```python
df_test_dummies
```

```python
df_test_dummies.reset_index(drop=True)
```

```python
def income(features, predict):
    count = 0
    for i in range(len(predict)):
        if (features.iloc[i].is_canceled == 1 and predict[i] == 1):
            count += features.iloc[i].profit * 0.2
        else:
            count += features.iloc[i].profit
    return count
```

```python
model = LogisticRegression(random_state=12345, class_weight='balanced', solver='lbfgs', max_iter=1000)
model.fit(features_train, target_train)
predict = model.predict(features_test)
print(f1_score(target_test, predict))

# на вменяемость

f1_0 = f1_score(target_test, [0 for i in range(len(target_test))])
f1_1 = f1_score(target_test, [1 for i in range(len(target_test))])
print(f1_0, f1_1)
```

```python
new_income = income(df_test_dummies.reset_index(drop=True), predict)
print('сколько поимеем в случае перехода на новую систему:', new_income - old_income)
```
